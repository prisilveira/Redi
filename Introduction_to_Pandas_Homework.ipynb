# Grouping

```python
# Get the average rent over all apartments
rentals['Rent'].mean()
```

```python
# Get the average rent per region (sort the results) and get the value for region 'Kreuzberg'.
# Turn the output into a Dataframe and the column which represents the average rent should be named 'avg_rent'
rentals.groupby('Region')['Rent'].mean().to_frame().rename(columns={'Rent':'avg_rent'})
```

```python
# Get the average rent grouped by region and condition. Try to use the different ways for doing aggregations as discussed during the class.
# What's the avg. rent for an apartment in 'Mitte' in 'well_kept' condition?
rentals.groupby(['Region','Condition'])['Rent'].mean()
```

```python
# Do different aggregations per (Region, Condition)
# 1. The average rent
rentals.groupby(['Region','Condition'])['Rent'].mean()
```

```python
# 2. The most recent construction year
rentals.groupby(['Region','Condition'])['Year_Construction'].max()
```

```python
# 3. The biggest apartment by space
rentals.groupby(['Region','Condition'])['Space'].max()
```

```python
# 4. The total number of rooms across all apartments
# The result should be a dataframe with 6 columns: 'Region', 'Condition', 'avg_rent', 'newest_year', 'biggest_apt', 'total_rooms'
rentals_grouped = rentals.groupby(['Region', 'Condition']).agg(
    avg_rent=('Rent', 'mean'),
    newest_year=('Year_Construction', 'max'),
    biggest_apt=('Space', 'max'),
    total_rooms=('Rooms', 'sum'))

print(rentals_grouped)
```

```python
# How many apartments are in Mitte in a 'well-kept' condition?

#first create a data_frame to save the numbers of apartment per condition category
filtered_mitte=rentals.groupby(['Region','Condition']).agg(number_apt=('Region','count'))
print(filtered_mitte)

#the output is a multiindex hierarquical dataset
```

```python
#second, convert the multiindex hierarquical dataset into a regular dataset and then filter the values that match Mitte and well_kept
# from the Region and Condition columns and then count the number of entries
filtered_mitte_modified=filtered_mitte.reset_index()
filtered_mitte_modified.head()

```

```python
#now filter values
filtered_mitte_modified[(filtered_mitte_modified['Region']=='Mitte') & (filtered_mitte_modified['Condition']=='well_kept')]

##function
filtered = filtered_mitte_modified[
    (filtered_mitte_modified['Region'] == 'Mitte') &
    (filtered_mitte_modified['Condition'] == 'well_kept')
]

if filtered.empty or filtered['number_apt'].values[0] < 1:
    print("There is no well-kept apartments in Mitte")
else:
    print(filtered)
```

```python
# How many apartments are there in Kreuzberg built before 1900?

#funtion
filtered = rentals[
    (rentals['Region'] == 'Kreuzberg') &
    (rentals['Year_Construction'] < 1900)
]

if filtered.empty:
    print("There is no apartment built before 1900 in Kreuzberg")
else:
    print(f"There are {len(filtered)} apartments built before 1900 in Kreuzberg.")
    print(filtered)
```

```python
# What are the average spaces of the apartments by the number of rooms?
average_space_by_rooms = rentals.groupby('Rooms')['Space'].mean()
print(average_space_by_rooms)
```

```python
# Which is the most expensive region on average and what is the maximum rent for this region?
average_rent_by_region = rentals.groupby('Region')['Rent'].mean()
print(average_rent_by_region)

#then we order in descending order
average_rent_by_region.sort_values(ascending=False)

def most_expensive_region(rentals):
    # Group by Region and calculate the average rent for each region but index must be reseted because it returns a multiindex hierarchical dataframe
    avg_rent_by_region = rentals.groupby('Region')['Rent'].mean().reset_index()

    # Sort the regions by average rent in descending order
    avg_rent_by_region_sorted = avg_rent_by_region.sort_values(by='Rent', ascending=False)

    # Get the region with the highest average rent (first in the sorted list)
    most_expensive = avg_rent_by_region_sorted.iloc[0]

    # Print the result
    print(f"The most expensive region is {most_expensive['Region']} with an average rent of {most_expensive['Rent']:.2f}.")

# Example usage
most_expensive_region(rentals)
```

```python
# What are the min, max and average number of rooms of the apartments in 'Köpenick'?
filtered_koepenick = rentals[rentals['Region'] == 'Köpenick']
min_rooms = filtered_koepenick['Rooms'].min()
max_rooms = filtered_koepenick['Rooms'].max()
avg_rooms = filtered_koepenick['Rooms'].mean()

############# function
def room_by_region(rentals, region_name):
  #first, the dataframe must befiltered for the specified region
  filtered_region=rentals[rentals['Region']== region_name]

  if filtered_region.empty:
    print(f"No data found for '{region_name}'.")
    return

  min_rooms=filtered_region['Rooms'].min()
  max_rooms=filtered_region['Rooms'].max()
  avg_rooms=filtered_region['Rooms'].mean()

  print(f"The minimum number of rooms in '{region_name}' is '{min_rooms}'.")
  print(f"The maximum number of rooms in '{region_name}' is '{max_rooms}'.")
  print(f"The average number of rooms in '{region_name}' is '{avg_rooms}'.")

###################### Usage
room_by_region(rentals,'Köpenick')
```

```python
import seaborn as sns

taxis = sns.load_dataset('taxis')
taxis.head()
```

```python
# Create a copy of the Dataframe and increase the 'fare' for any ride where the 'payment' is 'credit card' by 20% using .loc[]
taxis_cp=taxis.copy()
taxis_cp.head()
```

```python
#we create a new column called new_fare in the taxis_cp dataframe populated with the values from the column fare
taxis_cp['new_fare']=taxis_cp['fare']
taxis_cp.head()
```

```python
#now we create a condition where if payment == credit card, the value in fare should be multipled by 1.2 and stored in the respective
#cell of the column new_fare. But we have first to check how the value credit card is represented in the column payment
payment_type= taxis_cp['payment'].unique()
print(payment_type)
type(payment_type)
#get and store the position the position (index) of credit card in the array payment_type that has only on single element, that is, in position [0]
position_cc= np.where(payment_type=='credit card')[0]
print(position_cc) # this is a tuple and the position/index is stored as array[0] and not the actual number 0

#accessing the real position/index value
credit_card_position=payment_type[position_cc[0]]

#now that we now that credit card in in position 0 of the array position_cc, we can filter the dataframe to keep the rows of the taxi_cp of the columns
# payment and new_fare and update the values in the new_fare column
taxis_cp.loc[taxis_cp['payment']== credit_card_position, 'new_fare'] *= 1.2   #here we use loc to selct rows based on a labeled condition which is credit card in colum 'payment'
print(taxis_cp.head())
```
